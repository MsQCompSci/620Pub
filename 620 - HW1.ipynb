```python
import igraph as ig
```

```python
# Create a graph object
g = ig.Graph()

#print
print(g)
```

```python
#Define the node names and edge segments in a list
nodes = ['Andre', 'Beverly', 'Carol', 'Diane', 
        'Ed', 'Fernando', 'Heather', 'Ike', 'Jane','Garth']

#edge segments are in pairs - so list of tuples
edges = [('Andre', 'Beverly'),
         ('Andre', 'Carol'),
         ('Andre', 'Diane'),
         ('Andre', 'Fernando'),
         ('Beverly', 'Diane'),
         ('Beverly', 'Garth'),
         ('Beverly', 'Ed'),
         ('Carol', 'Fernando'),
         ('Carol', 'Diane'),
         ('Ed', 'Diane'),
         ('Ed', 'Garth'),
         ('Diane', 'Fernando'),
         ('Diane', 'Garth'),
         ('Fernando','Garth'),
         ('Heather', 'Fernando'),
         ('Heather', 'Garth'),
         ('Heather', 'Ike'),
         ('Ike','Jane' )]
```

```python
#add the nodes and vertices
g.add_vertices(nodes)
g.add_edges(edges)
```

```python
#summary
ig.summary(g)
## has 10 nodes and 18 edges
```

```python
#inspect the graph using degree with nodes to confirm number or associations
deg = g.degree()
for index in range(len(nodes)):
    print(nodes[index] + ':' + str(deg[index]))
```

```python
layout = g.layout("kk")
col = ['blue', 'red', 'green', 'yellow','orange', 'pink', 'purple', 'black', 'white', 'brown']
ig.plot(g, layout = layout, vertex_label= nodes, vertex_color= col, edge_width=[4], edge_color= 'grey')
```

Now lets try it using the NetworkX library

```python
#import network and plotting libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# greate a networkX object
G = nx.Graph()

#add the nodes and edges
G.add_nodes_from(nodes)
G.add_edges_from(edges)

#inspect using degree
print(nx.degree(G))
```

```python
#plot
nx.draw(G, with_labels = True)
plt.figure(figsize=(15,20))
plt.show()
```
